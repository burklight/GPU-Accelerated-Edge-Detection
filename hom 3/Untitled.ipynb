```python
import numpy as np
import matplotlib.pyplot as plt
dir_data = './Files/'
dir_imgs = './Images/'
```

```python
unordered = {}
ordered = {}
N = 100
P = 10
for i in range(P):
    unordered[i] = np.loadtxt(dir_data+'unord_'+str(i)+ "_" + str(N) + '.txt', delimiter=',')
    ordered[i] = np.loadtxt(dir_data+'ord_'+str(i)+"_" + str(N) +'.txt', delimiter=',')
```

```python
merged_unordered = []
merged_ordered = []
for i in range(P):
    merged_unordered = np.hstack((merged_unordered, unordered[i]))
    merged_ordered = np.hstack((merged_ordered, ordered[i]))
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True, )
ax[0].plot(merged_unordered, c="black", label = "Unordered")
ax[1].plot(merged_ordered, c="red", label = "Ordered")
ax[0].legend()
ax[1].legend()
#plt.savefig(dir_imgs+'proof_ord.eps', format='eps')
```
