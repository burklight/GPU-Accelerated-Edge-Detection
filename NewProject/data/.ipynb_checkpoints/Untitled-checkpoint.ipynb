```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import scipy.ndimage as sc

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
img = mpimg.imread('./512/img1.jpg')     
gray = rgb2gray(img).astype(int)   
plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
folders = ['./512/','./1024/','./2048/']
```

```python
dirImg = ''
dirTxt = ''
for folder in folders:
    for i in range(5):
        dirImg = folder + 'img' + str(i+1) + '.jpg'
        img = mpimg.imread(dirImg)     
        gray = rgb2gray(img).astype(int)   
        plt.imshow(gray, cmap = plt.get_cmap('gray'))
        plt.show()
        dirTxt = folder + 'img' + str(i+1) + '.txt'
        np.savetxt(fname=dirTxt, X=gray, delimiter=',',newline=',',fmt='%d')
```

```python
img = mpimg.imread('./512/img3.jpg')     
gray = rgb2gray(img).astype(int)
print(gray[5,5:10])
```

```python
b = np.array([[0, 1, 0],[1,-4,1],[0,1,0]])
g = np.array([[2,4,5,4,2],[4,9,12,9,4],[5,12,15,12,5],[4,9,12,9,4],[2,4,5,4,2]]) / 159
#b = np.array([[-1,-1,-1,],[-1,4,-1],[-1,-1,-1]])

c = np.zeros((512,512))
d = np.zeros((512,512))
sc.convolve(input=gray, weights=g, output=c)
c = c.astype(int)
sc.convolve(input=c, weights=b, output=d)
print(gray[5,5:10])
print(c[5,5:10])
print(d[5,5:10])

d = np.where(d>5,0,1)
```

```python
fig, ax = plt.subplots(figsize=(20,20))
ax.imshow(gray,cmap='gray')
fig, ax = plt.subplots(figsize=(20,20))
ax.imshow(d,cmap='binary')
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('./512/img3.jpg',0)
edges = cv2.Canny(img,100,200)
 
    
fig, ax = plt.subplots(figsize=(20,20))
ax.imshow(img,cmap='gray')
fig, ax = plt.subplots(figsize=(20,20))
ax.imshow(edges,cmap='gray') 
 
plt.show()
```
